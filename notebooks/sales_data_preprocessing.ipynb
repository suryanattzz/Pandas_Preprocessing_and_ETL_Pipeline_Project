```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', 20)
```

```python
df_trans = pd.read_csv(r"../data/raw/sales_data.csv")
df_trans
```

```python
df_trans.info()
```

## Null values

```python
df_trans.isnull().sum()        # column-wise count
# df_trans.isnull().sum().sum()  # total missing
```

```python
df_trans[['store_id','product_name','payment_mode','sales_rep_id']] = df_trans[['store_id','product_name','payment_mode','sales_rep_id']].fillna("unknown")
df_trans.isnull().sum() 
```

## Filling Unit_Price Values

```python
df_trans['unit_price'] = df_trans['unit_price'].fillna(df_trans['total_amount']/df_trans['quantity'])
```

```python
# df_trans['product_name'] = df_trans['product_name'].fillna(0)
# df_trans.isnull().sum() 
```

## Date Formatting

```python
df_trans['sale_date'] = pd.to_datetime(df_trans['sale_date'],format='mixed', errors='coerce')
df_trans
```

```python
df_trans.isnull().sum() 
```

## Remove Duplicate Values

```python
print(df_trans['transaction_id'].duplicated().sum())
print(df_trans['transaction_id'].value_counts())
```

```python
df_trans[df_trans['transaction_id'].duplicated(keep=False)].sort_values('transaction_id', ascending=False)
```

```python
df_trans.groupby('transaction_id').filter(lambda x: len(x) > 1).sort_values('transaction_id', ascending=False)
```

```python
df_trans = df_trans.drop_duplicates(subset=['transaction_id'], keep='first').reset_index(drop=True)
df_trans
```

```python
df_trans[df_trans['transaction_id'].duplicated(keep=False,)].sort_values('transaction_id', ascending=False)
```

## String Formatting

```python
df_trans['region'].unique()
```

```python
df_trans[['region','category','product_name']] = (
    df_trans[['region','category','product_name']].apply(lambda x: x.str.title() if isinstance(x, str) else x)
)
```
