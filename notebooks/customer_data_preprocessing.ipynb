## Customer Data

```python
df_cus.info()
```

### Validation of Mail and phone number

```python
# Phone validation
df_cus['phone_clean'] = df_cus['phone'].str.replace(r'[\s-]', '', regex=True)
phone_no_pattern = r'^(\+91)?[6-9]\d{9}$'
df_cus['valid_phone'] = df_cus['phone_clean'].str.match(
    phone_no_pattern,
    na=False
)
df_cus['valid_phone'].value_counts()
```

```python
email_pattern = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
df_cus['valid_email'] = df_cus['email'].str.match(
    email_pattern,
    na=False
)

df_cus['valid_email'].value_counts()
```

```python
df_cus[~df_cus['valid_email']]
```

```python
df_cus[~df_cus['valid_phone']]
```

```python
df_cus = df_cus.drop(['valid_phone', 'valid_email','phone_clean'], axis=1)
df_cus
```

## Null values

```python
df_cus[['email','phone','city','occupation','last_purchase_date']] = df_cus[['email','phone','city','occupation','last_purchase_date']].fillna('none')
df_cus.isnull().sum() 
```
