```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', None)
```

```python
df_ord = pd.read_csv(r"processed_orders_data.csv")
df_ord[:10]
```

```python
df_ord.info()
```

```python
df_prod = pd.read_json(r"products.json")
df_prod[0:10]
```

## 1) What is the total number of orders?

```python
total_orders = len(df_ord)

print(f"Total Orders: {total_orders:,}")
```

```python
df_ord['order_date'] = pd.to_datetime(
    df_ord['order_date'],
    errors='coerce'
)
```

```python
monthly_orders = (
    df_ord
    .groupby(df_ord['order_date'].dt.to_period('M')).size()
)

monthly_orders.plot(marker='o')

plt.title('Monthly Orders Trend')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.show()
```

```python
df_ord = df_ord.set_index('order_date')
daily_orders = df_ord.index.to_series().resample('D').count()

daily_orders.plot()
plt.title('Daily Orders')
plt.show()
```

## 2) Which cities placed the most orders?

```python
city_orders = (df_ord.groupby('customer_city').size().sort_values(ascending=False))

city_orders
```

```python
top_cities = (df_ord.groupby('customer_city').size().sort_values(ascending=False).head(10))

top_cities.plot(kind='bar')
plt.title('Top 10 Cities by Orders')
plt.ylabel('Total Orders')
plt.show()
```

## 3) What is the payment mode distribution?

```python
payment_dist = df_ord['payment_mode'].value_counts()

payment_dist
```

```python
payment_pct = (
    df_ord['payment_mode'].value_counts() / len(df_ord) * 100
)

print(f"Payment Mode Distribution: ")
print(payment_pct)
```

```python
payment_dist.plot(kind='bar')
plt.title('Payment Mode Distribution')
plt.xlabel('Payment Mode')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

## 4) What is the order status distribution?

```python
status_dist = df_ord['order_status'].value_counts()

status_dist
```

```python
status_pct = (
    df_ord['order_status'].value_counts() / len(df_ord) * 100
)

print(f"Order Status Distribution: ")
print(status_pct)
```

```python
status_dist.plot(kind='bar')
plt.title('Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

## 5) Which products were ordered most frequently?

```python
df_ord = df_ord.reset_index()
product_qty = df_ord['product_id'].value_counts().sort_values(ascending=False).head(10)

product_qty
```

```python
top_products = (df_ord['product_id'].value_counts().sort_values(ascending=False).head(10))

top_products.plot(kind='bar')
plt.title('Top 10 Products by Orders')
plt.ylabel('Number of Orders')
plt.show()
```

## 6) What is the total quantity sold?

```python
total_quantity = df_ord['quantity'].sum()

print(f"Total Quantity Sold: {total_quantity:,} units")
```

```python
avg_quantity = df_ord['quantity'].mean()

print(f"Average Quantity per Order: {avg_quantity:.2f} units")
```

## 7) What is the product catalog information?

```python
total_products = len(df_prod)

print(f"Total Products in Catalog: {total_products}")
```

```python
category_count = df_prod['category'].value_counts()

category_count
```

```python
category_count.plot(kind='bar')
plt.title('Products by Category')
plt.ylabel('Number of Products')
plt.show()
```

```python
min_price = df_prod['unit_price'].min()
max_price = df_prod['unit_price'].max()
avg_price = df_prod['unit_price'].mean()

print(f"Minimum Price: ₹{min_price:,.2f}")
print(f"Maximum Price: ₹{max_price:,.2f}")
print(f"Average Price: ₹{avg_price:,.2f}")
```
